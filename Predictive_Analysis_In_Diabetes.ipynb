## Outlier Detection

```python
# Splitting the data into input feature x (independent variable) and target value y (dependent variable)
x=df.drop(columns='Outcome',axis=1)
y=df['Outcome']
```

```python
fig,ax=plt.subplots(figsize=(15,15))
sns.boxplot(data=x,ax=ax)
```

```python
cols=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
for col in cols:
    Q1=x[col].quantile(0.25)
    Q3=x[col].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    mask=(x[col]>=lower_bound) & (x[col]<=upper_bound)

# Filter Dataset to remove outliers

X_outlier_detection=x[mask]        # Outlier Detected input Feature
Y_outlier_detection=y[mask]        # and corresponding target values
```

```python
X_outlier_detection.shape
```

```python
x.shape
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X_outlier_detection)
X_scaled=pd.DataFrame(X_scaled,columns=cols)
```

```python
X_scaled.reset_index(drop=True,inplace=True)
Y_outlier_detection.reset_index(drop=True,inplace=True)
```

```python
# Quantie Based Filtering
q=X_scaled['Insulin'].quantile(0.95)
mask=X_scaled['Insulin']<q
dataNew=X_scaled[mask]
Y_outlier_detection=Y_outlier_detection[mask]
```

```python
# Visualization with Box Plot
fig,ax=plt.subplots(figsize=(15,15))
sns.boxplot(data=dataNew,ax=ax)
```

```python
# Spilitting Data into training and testing sets
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(dataNew,Y_outlier_detection,test_size=0.33,random_state=42)
```

```python
## Shape After Data splitting 
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Handling Imbalanced Data
print(Y_train.value_counts())
```

```python
## Handling Imbalaced data through SMOTE
# Oversampling to minority class 
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
X_train_resampled, Y_train_resampled=smote.fit_resample(X_train,Y_train)

# Checking resample class distribution 
print("\nResampled class distribution:")
print(pd.Series(Y_train_resampled).value_counts())
```

## Model Training

```python
from sklearn.linear_model import LogisticRegression
logistic_regression_model=LogisticRegression()
logistic_regression_model.fit(X_train_resampled,Y_train_resampled)    ## fit the model to the training data
```

## Model Predictions

```python
y_predictions = logistic_regression_model.predict(X_test)
print(y_predictions)
```

## Model Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
accuracy_score(Y_test,y_predictions)

target_names=['Non-Diabetic','Diabetic']

# Confusion Matrix

plt.figure(figsize=(7,5))
sns.heatmap(confusion_matrix(Y_test, y_predictions), annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# Classification Report

print(classification_report(Y_test, y_predictions,target_names=target_names))
```
